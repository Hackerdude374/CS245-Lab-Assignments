```python
# Exercise 1: Read Total profit of all months and show it
# using a line plot
# Total profit data provided for each month. Generated line plot must include the following
# properties: –
# ● X label name = Month Number
# ● Y label name = Total profit
# The line plot graph should look like this.

import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Extracting data for plotting
months = df['month_number']
total_profit = df['total_profit']

# Creating the line plot
plt.plot(months, total_profit, marker='o', color='b', linestyle='-')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')

# Set tick locations and labels for x-axis
plt.xticks(range(1, 13, 1), rotation=45)

# Set tick locations and labels for y-axis
plt.yticks(range(100000, 600001, 100000))

# Displaying the plot
plt.grid(True)           # Add gridlines
plt.tight_layout()       # Adjust layout to prevent clipping of labels
plt.show()
```

```python

# Exercise 3: Read all product sales data and show it  using a multiline plot
# Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
# The graph should look like this.

 
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Extracting data for plotting
months = df['month_number']
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

# Creating the multiline plot for each product
plt.plot(months, facecream, marker='o', label='Face Cream')
plt.plot(months, facewash, marker='o', label='Face Wash')
plt.plot(months, toothpaste, marker='o', label='Toothpaste')
plt.plot(months, bathingsoap, marker='o', label='Bathing Soap')
plt.plot(months, shampoo, marker='o', label='Shampoo')
plt.plot(months, moisturizer, marker='o', label='Moisturizer')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Product Sales Data')

# Show legend at the lower center location
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3)

# Set tick locations and labels for x-axis
plt.xticks(range(1, 13, 1), rotation=45)

# Displaying the plot
plt.grid(True)           # Add gridlines
plt.tight_layout()       # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
# The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.
# The bar chart should look like this.

 
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Extracting data for plotting
months = df['month_number']
facecream_sales = df['facecream']
facewash_sales = df['facewash']

# Creating the bar chart
plt.bar(months - 0.2, facecream_sales, width=0.4, label='Face Cream', align='center', color='blue')
plt.bar(months + 0.2, facewash_sales, width=0.4, label='Face Wash', align='center', color='orange')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Face Cream and Face Wash Sales per Month')

# Adding legend
plt.legend()

# Displaying the plot
plt.xticks(months)
plt.grid(True)           # Add gridlines
plt.tight_layout()       # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
# The histogram should look like this.
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Extracting data for plotting
total_profit = df['total_profit']

# Creating the histogram
plt.hist(total_profit, bins=10, color='blue', edgecolor='black')

# Adding labels and title
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Total Profit Distribution')

# Set tick locations and labels for x-axis
x_ticks = list(range(150000, 375001, 25000))
x_ticks[4] = 250000  # Change tick at 250000 to 50k interval
plt.xticks(x_ticks, rotation=45)

# Set tick locations and labels for y-axis
plt.yticks(range(0, 6, 1))

# Displaying the plot
plt.grid(True)           # Add gridlines
plt.tight_layout()       # Adjust layout to prevent clipping of labels
plt.show()
```

```python

# Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
# Note: In Pie chart display Number of units sold per year for each product in percentage.
# The Pie chart should look like this.

 
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Calculate total sale data for last year for each product
total_units_sold_last_year = df.groupby('month_number').sum().iloc[-12:, 1:-2].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_units_sold_last_year, labels=total_units_sold_last_year.index, autopct='%1.1f%%', startangle=140)

# Adding title
plt.title('Total Sale Data for Last Year (Per Product)')

# Displaying the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
 
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
# The Subplot should look like this.

import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Extracting data for plotting
months = df['month_number']
bathingsoap_sales = df['bathingsoap']
facewash_sales = df['facewash']

# Creating subplots
fig, axs = plt.subplots(2)

# Plotting Bathing Soap sales
axs[0].plot(months, bathingsoap_sales, marker='o', color='black')
axs[0].set_title('Bathing Soap Sales')

# Plotting Facewash sales
axs[1].plot(months, facewash_sales, marker='o', color='red')
axs[1].set_title('Facewash Sales')

# Setting common labels
for ax in axs:
    ax.set_xlabel('Month Number')
    ax.set_ylabel('Number of Units Sold')

# Adjust layout
plt.tight_layout()

# Displaying the plot
plt.show()

 
```

```python

# Exercise Question 10: Read all product sales data and show it using the stack plot
# The Stack plot should look like this.
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# Extracting data for plotting
months = df['month_number']
facecream_sales = df['facecream']
facewash_sales = df['facewash']
toothpaste_sales = df['toothpaste']
bathingsoap_sales = df['bathingsoap']
shampoo_sales = df['shampoo']
moisturizer_sales = df['moisturizer']

# Creating the stack plot
plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales,
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'],
              colors=['purple', 'turquoise', 'red', 'black', 'green', 'yellow'])

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Product Sales Data')

# Adding legend
plt.legend(loc='upper left')

# Displaying the plot
plt.tight_layout()
plt.show()
```
