```python
# Exercise 1: Read Total profit of all months and show it
# using a line plot
# Total profit data provided for each month. Generated line plot must include the following
# properties: –
# ● X label name = Month Number
# ● Y label name = Total profit
# The line plot graph should look like this.import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

months = df['month_number']
total_profit = df['total_profit']

plt.plot(months, total_profit, marker='o', color='b', linestyle='-')
# ● X label name = Month Number
# ● Y label name = Total profit
plt.xlabel('Month Number') 
plt.ylabel('Total Profit')

plt.title('Total Profit of All Months')
plt.xticks(range(1, 13, 1), rotation=45)
plt.yticks(range(100000, 600001, 100000))
plt.grid(True)          
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Exercise 2: Get total profit of all months and show line plot with the following Style properties
# Generated line plot must include following Style properties: –
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3
# The line plot graph should look like this.

months = df['month_number']
sold_units = df['total_units']

# Generated line plot must include following Style properties: –
# Line Style dotted and Line-color should be red
# Add a circle marker.
# Line marker color as read
# Line width should be 3
plt.plot(months, sold_units, linestyle='dotted', color='red', marker='o', markersize=8, markerfacecolor='red', linewidth=3, label='Sold Units')
plt.xlabel('Month Number')  # X label name = Month Number
plt.ylabel('Sold Units Number')  # Y label name = Sold units number
plt.title('Sold Units Number of All Months')  # Title

# Show legend at the lower right location.
plt.legend(loc='lower right')

plt.xticks(range(1, 13, 1), rotation=45)

plt.grid(True)
plt.show()
```

```python

# Exercise 3: Read all product sales data and show it  using a multiline plot
# Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
# The graph should look like this.
import matplotlib.pyplot as plt
import pandas as pd

# multiline plot
# Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
months = df['month_number']
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    plt.plot(months, df[product], marker='o', label=product.capitalize())
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.title('Product Sales per Month')

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3)
plt.xticks(range(1, 13), rotation=45)
plt.grid(True)

plt.show()
```

```python
 
# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
# Also, add a grid in the plot. gridline style should “–“.
# The scatter plot should look like this.

 
 
import matplotlib.pyplot as plt
import pandas as pd
# Read toothpaste sales data of each month and show it using a scatter plot
months = df['month_number']
toothpaste_sales = df['toothpaste']

plt.scatter(months, toothpaste_sales, color='blue', marker='o')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales per Month')

#  Also, add a grid in the plot. gridline style should “–“.
plt.grid(True, linestyle='--')
plt.show()
```

```python
# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
# The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.
# The bar chart should look like this.
import matplotlib.pyplot as plt
import pandas as pd

# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
m = df['month_number']
fc = df['facecream']
fw = df['facewash']
# The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.
plt.bar(m - 0.2, fc,   width=0.4, label='A',align='center', color='b')
plt.bar(m + 0.2, fw,  width=0.4, label='B',  align='center', color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Z')
plt.legend()
plt.xticks(m)
plt.grid(True)
plt.show()
```

```python
# Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges
# The histogram should look like this.
import matplotlib.pyplot as plt
import pandas as pd

total_profit = df['total_profit']

# histogram
plt.hist(total_profit, bins=10, color='blue', edgecolor='black', alpha = 0.7)
#  most common profit ranges
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Total Profit Distribution')
#ranges
x_ticks = list(range(150000, 375001, 25000))
x_ticks[4] = 250000  # Change tick at 250000 to 50k interval
plt.xticks(x_ticks, rotation=45)


plt.yticks(range(0, 6, 1))
plt.grid(True)       
plt.show()
```

```python

# Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
# Note: In Pie chart display Number of units sold per year for each product in percentage.
# The Pie chart should look like this.
import matplotlib.pyplot as plt
import pandas as pd

# Calculate total sale data for last year for each product and show it using a Pie chart
total_units_sold_last_year = df.groupby('month_number').sum().iloc[-12:, 1:-2].sum() # 12 months (12 rows)

# Pie chart creation
plt.figure(figsize=(8, 8))
plt.pie(total_units_sold_last_year, labels=total_units_sold_last_year.index, autopct='%1.1f%%', startangle=140)


plt.title('Total Sale Data for Last Year Percentage')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
 
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot
# The Subplot should look like this.

import matplotlib.pyplot as plt
import pandas as pd
months = df['month_number']
bathingsoap_sales = df['bathingsoap']
facewash_sales = df['facewash']

# create a subPlot
fig, axs = plt.subplots(2)

axs[0].plot(months, bathingsoap_sales, marker='o', color='black')
axs[0].set_title('Bathing Soap Sales')


axs[1].plot(months, facewash_sales, marker='o', color='red')
axs[1].set_title('Facewash Sales')

for ax in axs:
    ax.set_xlabel('Month Number')
    ax.set_ylabel('Number of Units Sold')

plt.show()

 
```

```python

# Exercise Question 10: Read all product sales data and show it using the stack plot
# The Stack plot should look like this.
import matplotlib.pyplot as plt
import pandas as pd
months = df['month_number']
facecream_sales = df['facecream']
facewash_sales = df['facewash']
toothpaste_sales = df['toothpaste']
bathingsoap_sales = df['bathingsoap']
shampoo_sales = df['shampoo']
moisturizer_sales = df['moisturizer']
plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales,
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'],
              colors=['purple', 'turquoise', 'red', 'black', 'green', 'yellow'])


plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Product Sales Data')
plt.legend(loc='upper left')
plt.show()
```
