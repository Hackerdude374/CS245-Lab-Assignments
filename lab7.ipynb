```python

# Question 3

print("Question 3: Split the DataFrame into X (the data) and y (the labels). This function should return a tuple of length 2: (X, y), where X, a pandas DataFrame, has shape (569, 30) y, a pandas Series, has shape (569,).")
#split dataframe into 
# X: data
# y: labels
def splitDataFrame():
    x = pd.DataFrame( cancer.data, columns = cancer.feature_names) # data frame
    
    y = pd.Series(cancer.target)
    
    return x, y

x,y = splitDataFrame()
# shape (569, 30) y, a pandas Series, has shape (569,).")
print("x shape: ", x.shape)
print("y shape: ", y.shape)
print("------------------------------------------------------------------------------------------------")
```

```python
# Question 4
from sklearn.model_selection import train_test_split
# Using train_test_split, split X and y into training and test sets (X_train, X_test, y_train, and y_test).
print("Question 4: Using train_test_split, split X and y into training and test sets (X_train, X_test, y_train, and y_test)")

# (X_train, X_test, y_train, and y_test).
def TrainingAndTestSets(x, y):
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) #0.2 20% data testing, 42 means same random splits every tme code generated
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = TrainingAndTestSets(x, y)
print("x_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("x_test:", X_test)
print("y_test:", y_test)
print("------------------------------------------------------------------------------------------------")
```

```python
# Question 5 
from sklearn.neighbors import KNeighborsClassifier
# Using KNeighborsClassifier, fit a k-nearest neighbors (knn) classifier with X_train, y_train and using one nearest neighbor (n_neighbors = 1). This function should return a sklearn.neighbors.classification.KNeighborsClassifier.
print("Question 5:  Using KNeighborsClassifier, fit a k-nearest neighbors (knn) classifier with X_train, y_train and using one nearest neighbor (n_neighbors = 1). This function should return a sklearn.neighbors.classification.KNeighborsClassifier.)")
# Using KNeighborsClassifier, fit a k-nearest neighbors (knn) classifier with X_train, y_train and using one nearest neighbor (n_neighbors = 1). 
# This function should return a sklearn.neighbors.classification.KNeighborsClassifier.
def classifier(X_train, y_train):
    classifier = KNeighborsClassifier(n_neighbors=1)  # 1 = nearest neighor
    classifier.fit(X_train, y_train)  #fit them
    return classifier

# Call the function to fit the knn classifier
kNeighorclassifier = classifier(X_train, y_train)

# Print the knn classifier
print("knn classifier:", kNeighorclassifier)
print("------------------------------------------------------------------------------------------------")
```

```python
print("Question 6 Using your knn classifier, predict the class label using the mean value for each feature. Hint: You can use cancerdf.mean()[:-1].values.reshape(1, -1) which gets the mean value for each feature, ignores the target column, and reshapes the data from 1 dimension to 2 (necessary for the precict method of KNeighborsClassifier). This function should return a numpy array either array([ 0.]) or array([ 1.])");

# Question 6 Using your knn classifier, predict the class label using the mean value for each feature. Hint: You can use cancerdf.mean()[:-1].values.reshape(1, -1)
# which gets the mean value for each feature, ignores the target column, and reshapes the data from 1 dimension to 2 (necessary for the precict method of KNeighborsClassifier). 
# This function should return a numpy array either array([ 0.]) or array([ 1.])");
def predict_with_mean(kNeighorclassifier):
    cancerdf = pd.DataFrame(data=cancer.data, columns=cancer.feature_names) #removed -1
    featureMeans = cancerdf.mean().values.reshape(1, -1)
    predictions = kNeighorclassifier.predict(featureMeans)
    return predictions

predictionWithMean = predict_with_mean(kNeighorclassifier)

print("prediction using mean valuies: ", predictionWithMean)
print("------------------------------------------------------------------------------------------------")
```

```python
print("Question 7 Using your knn classifier, predict the class labels for the test set X_test. This function should return a numpy array with shape (143,) and values either 0.0 or 1.0")
def predict_labels(kNeighorclassifier, X_test):
    return kNeighorclassifier.predict(X_test)
# This function should return a numpy array with shape (143,) and values either 0.0 or 1.0"
predicted_labels = predict_labels(kNeighorclassifier, X_test)
print("predicted_labels", predicted_labels)
print("Shape", predicted_labels.shape)
```

```python

print("Question 8 Find the score (mean accuracy) of your knn classifier using X_test and y_test. This function should return a float between 0 and 1")
def knn_classifier_score(kNeighorclassifier, X_test, y_test):
    return kNeighorclassifier.score(X_test, y_test)

meanaccuracyscore = knn_classifier_score(kNeighorclassifier, X_test, y_test)
print("meanaccuracyscore: ", meanaccuracyscore)

print("------------------------------------------------------------------------------------------------")
```
